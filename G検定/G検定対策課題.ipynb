#### 問５　転移学習とは何か？以下のセルに記入せよ 

- 転移学習(Transfer Learning)とは、ある領域で学習したこと(学習済みモデル)を別の領域に役立たせ、効率的に学習させる方法です。
    
-  転移学習の主な特徴や利点は以下の通りです：

データ効率性: 新しいタスクに対してゼロからモデルを訓練する代わりに、既存の学習済みモデルを利用するため、データの量を大幅に削減できます。

高性能: 一般的な特徴や知識を持つ学習済みモデルを使用するため、新しいタスクにおいても高い性能を達成できる可能性が高いです。

転移の柔軟性: さまざまなタスクに対して転移学習を適用できます。例えば、画像認識から自然言語処理への転移など、異なるドメインやタスク間での適用が可能です。

時間とリソースの節約: 転移学習は、新しいタスクにモデルを再訓練する手間や計算リソースを節約できます。

#### 問８　強いAIと弱いAIの違いは何か？以下のセルに記入せよ

　　強いAIは一般知性を持ち、幅広い知識と認識能力を持つことを目指す一方、弱いAIは特定のタスクに特化した限定的な知能を持つものです。現在のAI技術は主に弱いAIに該当し、強いAIやAGIの実現にはまだ多くの研究と開発が必要です。
  
 -  強いAI（Strong AI）：

強いAIは、人間の知識、認識、思考、判断などの能力を模倣または再現することを目指すAIです。
これは一般的な人間の知能を持ち、意識や自己認識の能力を含むとされることがあります。つまり、強いAIは「意識的な知性」を持つことが仮定されます。
強いAIは自己学習し、問題を解決し、人間と同じように抽象的な理解や一般的な推論を行えると考えられますが、現実の世界でまだ達成されていない段階です。これは一種の人工一般知性（AGI）を目指すものです。

- 弱いAI（Weak AI）：

弱いAIは、特定のタスクやドメインに特化したAIで、限定的な知識やスキルを持つことが特徴です。
このタイプのAIは、特定の問題を解決するためにプログラムされたり、訓練されたりしますが、一般的な知識や認識能力を持たず、他のタスクに適用することはできません。
弱いAIは、音声認識、画像分類、自然言語処理、ゲームプレイなどの特定のタスクに利用されます。
　　  
　　

####  問９　ニューラルネットワークの隠れ層を増やすことによって、どんな問題が発生するか？以下のセルに記入せよ

　　ニューラルネットワークの隠れ層を増やすことによって、以下のような問題が発生する可能性があります：
  
  - 過学習（Overfitting）: 隠れ層の数を増やすと、モデルは訓練データに対して適応しすぎる可能性が高まります。これは、訓練データに含まれるノイズや局所的な特徴までを学習し、新しいデータに対する汎化性能が低下することを意味します。過学習は、モデルの複雑さが高まるにつれて一般的になります。

- 計算コストの増加: 隠れ層の数を増やすと、ネットワークのパラメータ数が増え、計算コストが増加します。特に大規模なニューラルネットワークでは、訓練と推論にかかる時間や計算リソースが増大する可能性があります。

- 勾配消失問題: 隠れ層を増やすと、ネットワークが非常に深くなり、勾配消失問題が顕著になる可能性があります。勾配消失問題は、逆伝播中に勾配が極端に小さくなり、訓練が収束しづらくなる現象です。

- ハイパーパラメータの調整の難しさ: 隠れ層の数を増やすと、ネットワークのアーキテクチャに関するハイパーパラメータの調整が難しくなります。適切な隠れ層の数、ユニット数、学習率、正則化などを設定する必要があり、その調整には多くの試行とエクスペリメントが必要です。

- データの量と品質の要求: 隠れ層を増やすと、訓練データの量と品質がより重要になります。十分なデータがない場合、過学習のリスクが高まります。
　　
　　

#### 問１０　正解率とはどうのような指標か？定義を述べよ

　　正解率は、モデルが正しく分類したサンプルの数を、全体のサンプル数で割って計算します。具体的な数式は以下の通りです：

- 正解率 (Accuracy)= 
正しく分類されたサンプル数/
全体のサンプル数

 
- 正解率は通常、0から1の値を取り、1に近いほどモデルの性能が高いことを示します。この指標は、モデルが予測したクラスが実際のクラスと一致する割合を示すため、一般的にモデルの性能を評価するのに役立ちます。

- ただし、正解率はすべてのクラスに対して均等に評価されるため、クラス間の不均衡がある場合には注意が必要です。特に、クラスの分布が不均衡である場合、正解率だけではモデルの性能を正確に評価できないことがあります。このような場合、適切な評価指標（例：適合率、再現率、F1スコアなど）を使用することが推奨されます。
